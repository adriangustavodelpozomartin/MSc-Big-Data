# Introduction to Bayesian Statistics

In this notebook, we will explore the fundamentals of Bayesian statistics, a powerful framework that allows us to update our beliefs as new information becomes available. Bayesian statistics is widely used in fields like machine learning, data science, and artificial intelligence, providing a flexible and interpretable approach to inference.

Bayesian inference differs significantly from frequentist statistics. While frequentist methods rely solely on data from experiments or observations, Bayesian statistics allows us to incorporate prior knowledge or beliefs into the analysis.

Throughout this notebook, we will:
- Compare Bayesian and frequentist approaches,
- Understand key concepts in Bayesian inference, and
- Develop a foundation for working with prior and posterior distributions.

## What is Bayesian Statistics?

Bayesian statistics is based on the idea of updating probabilities as new evidence is introduced. This approach is grounded in **Bayes' Theorem**, which provides a mathematical rule to calculate how likely a hypothesis is, given prior knowledge and current evidence.

The main elements of Bayesian inference include:
- **Prior Probability**: Our initial belief about the parameter before seeing any data.
- **Likelihood**: The probability of observing the data given the parameter.
- **Posterior Probability**: The updated belief about the parameter after observing the data.

In simple terms, Bayesian inference is a way of combining our existing knowledge (prior) with new information (likelihood) to make better-informed conclusions (posterior).


### A brief history of Bayesian statistics
Bayesian statistics has a rich history that dates back to the 18th century. The foundation was laid by Reverend Thomas Bayes (1701–1761), who introduced a method to update probabilities based on new evidence. His work, published posthumously in 1763, presented what is now known as Bayes' Theorem. 

![bayes_theorem_modified](bayes_face.webp){align="center"}

In the early 19th century, French mathematician Pierre-Simon Laplace (1749–1827) expanded upon Bayes' ideas, developing a comprehensive framework for Bayesian inference. Laplace applied these principles to various scientific problems, including celestial mechanics and medical statistics. 

Despite its early development, Bayesian methods faced criticism and were overshadowed by frequentist approaches in the early 20th century. However, the mid-20th century saw a resurgence of interest, notably through the work of Sir Harold Jeffreys, who advocated for Bayesian methods in scientific inference. 

The advent of modern computing in the late 20th century significantly advanced Bayesian statistics. Techniques such as Markov Chain Monte Carlo (MCMC) made complex Bayesian computations feasible, leading to widespread adoption in fields like machine learning, data science, and artificial intelligence. 


## Types of Uncertainty (pregunta de examen)
In Bayesian analysis, we encounter two primary types of uncertainty:

- **Aleatoric Uncertainty**: This represents randomness inherent to the system, which cannot be reduced by gathering more data. For example, the randomness in rolling a fair die is an aleatoric uncertainty.

- **Epistemic Uncertainty**: This type of uncertainty arises from a lack of knowledge about the system. Unlike aleatoric uncertainty, epistemic uncertainty can be reduced by gathering additional information. For example, our uncertainty about the fairness of a coin can be reduced by observing more coin flips.

Bayesian statistics provides a structured way to update epistemic uncertainty using new data, allowing us to form a more refined view of the underlying system.

# Basic Concepts of Bayesian Statistics

To understand Bayesian inference, let's go over some key terms and concepts.

### Key Terms
1. **Prior (Prior Probability)**: Represents our initial belief about the parameter before we observe any data.
2. **Likelihood**: The probability of observing the data, given the parameter value. It indicates how likely it is to observe the data if our hypothesis about the parameter is true.
3. **Posterior (Posterior Probability)**: The updated belief about the parameter after observing the data. The posterior combines the information from the prior and the likelihood.
4. **Evidence**: The total probability of observing the data under all possible hypotheses. It acts as a normalizing factor in Bayes' Theorem.

These concepts work together to help us update our beliefs in a logical way. In Bayesian inference, we start with a prior, gather evidence from data, and combine these to obtain a posterior.

Now, let's take a closer look at Bayes' Theorem.

## Bayes' Theorem

Bayes' Theorem allows us to calculate the posterior probability given a prior and a likelihood. Mathematically, it can be expressed as:

$$
P(\theta | D) = \frac{P(D | \theta) \cdot P(\theta)}{P(D)}
$$

where:
- $P(\theta | D)$ is the **posterior probability**, representing our updated belief about the parameter $\theta$ after observing the data $D$.
- $P(D | \theta)$ is the **likelihood**, or the probability of observing the data given a specific value of $\theta$.
- $P(\theta)$ is the **prior probability**, our initial belief about $\theta$.
- $P(D)$ is the **evidence**, the total probability of observing the data.

Bayes' Theorem helps us balance prior beliefs and new data to form an updated view.

![bayes_theorem_modified](xkcd_bayes.webp){align="center"} 




To derive the posterior distribution, let’s break down each component:

1. **Prior Distribution** $P(\theta)$: 
   This is our initial belief about the distribution of the parameter $\theta$ before observing any data. We choose the prior based on our existing knowledge or assumptions about $\theta$.

2. **Likelihood** $P(D | \theta)$: 
   This is the probability of observing the data $D$ given a specific value of $\theta$. The likelihood function quantifies how well different values of $\theta$ explain the observed data.

3. **Evidence** $P(D)$: 
   The evidence is the probability of observing the data $D$ under all possible values of $\theta$:
   $$
   P(D) = \int P(D | \theta) P(\theta) \, d\theta
   $$
   This integral sums up the contributions from all possible parameter values, making $P(D)$ a normalizing factor that ensures $P(\theta | D)$ is a valid probability distribution.

4. **Posterior Distribution** $P(\theta | D)$: 
   By combining the prior and likelihood, we obtain the posterior distribution:
   $$
   P(\theta | D) = \frac{P(D | \theta) \cdot P(\theta)}{P(D)}
   $$
   Substituting the evidence from step 3, we get:
   $$
   P(\theta | D) = \frac{P(D | \theta) \cdot P(\theta)}{\int P(D | \theta) P(\theta) \, d\theta}
   $$
   This is the fundamental equation in Bayesian inference. It shows that the posterior distribution is proportional to the product of the likelihood and the prior:
   $$
   P(\theta | D) \propto P(D | \theta) \cdot P(\theta)
   $$

This relationship allows us to calculate the posterior distribution as long as we know the prior and likelihood. However, we do not like to compute integrals, so we would use a trick called conjugation.


### Estimating the Parameter $\theta$ from the Posterior

Once we have obtained the posterior distribution $P(\theta | D)$, we can use it to estimate the parameter $\theta$. Bayesian inference allows us to consider the full distribution of $\theta$, providing several ways to summarize or estimate the parameter based on the posterior:

1. **Posterior Mean**:
   The posterior mean is the expected value of $\theta$ given the data, representing the "average" value of $\theta$ according to the posterior distribution:
   $$
   \mathbb{E}[\theta | D] = \int \theta \, P(\theta | D) \, d\theta
   $$
   The posterior mean is often used as a point estimate because it minimizes the expected squared error loss.

2. **Posterior Mode (Maximum A Posteriori or MAP Estimate)**:
   The posterior mode, or **MAP estimate**, is the value of $\theta$ that maximizes the posterior distribution:
   $$
   \theta_{\text{MAP}} = \arg \max_{\theta} P(\theta | D)
   $$
   The MAP estimate is useful when we want the most probable value of $\theta$ given the data.

3. **Posterior Median**:
   The posterior median is the value of $\theta$ such that half of the posterior probability mass lies below it and half lies above it:
   $$
   P(\theta \leq \text{median} | D) = 0.5
   $$
   The posterior median is a robust measure of central tendency, especially for skewed distributions.

4. **Credible Intervals**:
   Instead of a single point estimate, we can provide a range that captures a specified probability mass (e.g., 95%) of the posterior distribution. This interval, known as a **credible interval**, is given by:
   $$
   \text{Credible Interval} = [\theta_{\text{lower}}, \theta_{\text{upper}}]
   $$
   where $P(\theta_{\text{lower}} \leq \theta \leq \theta_{\text{upper}} | D) = 0.95$ for a 95% credible interval. The credible interval provides an estimate of $\theta$ with a specified level of confidence.

## Model vs. Likelihood in Bayesian Inference

In Bayesian inference, understanding the distinction between the **model** and the **likelihood** is crucial:

- **Model (Aleatoric Uncertainty)**: The model, expressed as $p(y | \theta)$, represents the probability of observing data $y$ given a fixed parameter $\theta$. It captures aleatoric uncertainty—the natural variability in the data.

- **Likelihood (Epistemic Uncertainty)**: The likelihood function, also denoted $p(y | \theta)$ but interpreted differently, gives us information about $\theta$ based on observed data $y$. It is not a probability distribution for $\theta$ but provides a way to update our beliefs (epistemic uncertainty) about $\theta$ when combined with the prior.

In essence, Bayesian inference combines the likelihood and prior uncertainty $p(\theta)$ to produce an updated posterior uncertainty $p(\theta | y)$, balancing both aleatoric and epistemic uncertainties.


# Bayesian vs. Frequentist Approach

To better understand Bayesian statistics, let's compare it with the frequentist approach through a simple example. Suppose we are trying to estimate the probability of a coin landing heads.

- In the **frequentist approach**, we calculate the probability based on observed data only.
- In the **Bayesian approach**, we incorporate prior knowledge or beliefs and update them as we observe new data.

## Problem Setup

Imagine we have a coin, and we want to estimate the probability of it landing heads, denoted by $p$. We’ll approach this problem by simulating 20 coin flips and observing the outcomes.
- **Frequentist Approach**: We’ll use the proportion of heads in the observed data as the estimate for $p$.
- **Bayesian Approach**: We’ll start with a prior belief about $p$ (e.g., a Beta distribution) and update this belief as we see more data.



```python
# Simulating data: 20 coin flips with a true probability of heads of 0.6
np.random.seed(42)
true_p = 0.6  # True probability of heads
flips = np.random.binomial(1, true_p, 20)  # 20 coin flips
observed_heads = np.sum(flips)

# Display results
print(f"Out of 20 coin flips, we observed {observed_heads} heads.")
```

- Frequentist Approach: Estimating the Probability of Heads

In the frequentist approach, we estimate the probability of heads simply by taking the proportion of heads in the observed data. This approach does not use prior knowledge about the coin; it solely relies on the observed data.


```python
# Frequentist estimate of probability
frequentist_p = observed_heads / len(flips)
print(f"Frequentist estimate of p (probability of heads): {frequentist_p:.2f}")
```

- Bayesian Approach: Estimating the Probability of Heads

In the Bayesian approach, we start with a **prior distribution** representing our belief about the probability of heads before seeing any data. For this example, let's assume a **Beta distribution** as our prior:

$$
\text{Beta}(\theta | \alpha, \beta) = \frac{\theta^{\alpha - 1} (1 - \theta)^{\beta - 1}}{B(\alpha, \beta)}
$$

where:
- $\theta$ is the probability of success,
- $\alpha$ and $\beta$ are the shape parameters, and
- $B(\alpha, \beta)$ is the **Beta function**, defined as:

$$
B(\alpha, \beta) = \int_0^1 t^{\alpha - 1} (1 - t)^{\beta - 1} \, dt
$$

The Beta distribution is a common choice for prior probabilities of discrete events because it is a conjugate of the Binomial distribution. We will choose a prior with parameters $\alpha = 2$ and $\beta = 2$, reflecting a neutral belief about $p$.


```python
from scipy.stats import beta

# Define the prior: Beta distribution with alpha=2, beta=2
alpha_prior = 2
beta_prior = 2

# Plotting the prior distribution
x = np.linspace(0, 1, 1000)
prior_dist = beta.pdf(x, alpha_prior, beta_prior)

plt.plot(x, prior_dist, label="Prior (Beta(2, 2))")
plt.xlabel("Probability of Heads (p)")
plt.ylabel("Density")
plt.title("Prior Distribution")
plt.legend()
plt.show()
```

Now that we have defined our prior belief about the probability of heads (using a Beta distribution), let's see how we can update this belief as we collect new data. 

In this example, we update our belief about the coin's probability of landing heads. For this, we need a likelihood function.

The **likelihood function** measures how well different values of $\theta$ (the probability of heads) explain the observed data. For a sequence of coin tosses, this likelihood function is based on the binomial distribution. The **Binomial distribution** models the probability of observing a specific number of successes in a fixed number of independent trials, where each trial has the same probability of success. The probability mass function of the Binomial distribution is given by:

$$
\text{Binomial}(k | n, \theta) = \binom{n}{k} \theta^k (1 - \theta)^{n - k}
$$

where:
- $k$ is the number of successes,
- $n$ is the number of trials,
- $\theta$ is the probability of success on each trial, and
- $\binom{n}{k}$ is the binomial coefficient, calculated as:

$$
\binom{n}{k} = \frac{n!}{k! (n - k)!}
$$


Now, let's update our prior distribution through the likelihood function with the observed data to obtain the **posterior distribution**. The posterior combines our prior belief with the likelihood of the observed data, allowing us to make an updated estimate of $p$.

For a Beta prior and binomial likelihood, the posterior distribution also follows a Beta distribution with parameters:
- $\alpha_{posterior} = \alpha_{prior} + k$
- $\beta_{posterior} = \beta_{prior} + (n - k)$

In our example, $n$ is the total number of tosses and $k$ is the total number of heads observed (and $n-k$ is the total number of tails observed).


```python
from scipy.stats import beta

# Updating the prior with observed data
alpha_post = alpha_prior + observed_heads
beta_post = beta_prior + len(flips) - observed_heads

# Posterior distribution
posterior_dist = beta.pdf(x, alpha_post, beta_post)

# Plotting prior and posterior distributions
plt.plot(x, prior_dist, label="Prior (Beta(2, 2))", color="blue", alpha=0.3)
plt.plot(x, posterior_dist, label="Posterior (updated with data)", color="orange")
plt.xlabel("Probability of heads (θ)")
plt.ylabel("Density")
plt.title("Posterior Distribution after Observing Data")
plt.legend()
plt.show()

# Bayesian estimate (mean of the posterior distribution)
bayesian_p = alpha_post / (alpha_post + beta_post)
print(f"Bayesian estimate of p (posterior mean): {bayesian_p:.2f}")

# 95% Credible Interval (confidence interval for Bayesian estimation)
credible_interval = beta.ppf([0.025, 0.975], alpha_post, beta_post)
print(f"95% credible interval for p: [{credible_interval[0]:.2f}, {credible_interval[1]:.2f}]")
```

### The Effect of Sample Size

Let's simulate a series of coin toss experiments with different sample sizes and observe the evolution of the posterior distribution.

```python
# Defining a function to update the posterior based on observed data
def update_posterior(alpha_prior, beta_prior, heads, tails):
    posterior_alpha = alpha_prior + heads
    posterior_beta = beta_prior + tails
    posterior_dist = beta(posterior_alpha, posterior_beta)
    mean = posterior_alpha / (posterior_alpha + posterior_beta)  # Calculate the mean of the posterior
    credible_interval = beta.ppf([0.025, 0.975], posterior_alpha, posterior_beta)
    return posterior_dist.pdf(x), mean, credible_interval

# Sample sizes to test (e.g., 10, 50, 100, 1000 coin flips)
sample_sizes = [10, 50, 100, 1000]
posteriors = []
means = []  # List to store means of each posterior
credible_intervals = []  # List to store credible interval of each posterior

# Define prior parameters correctly
alpha_prior = 2
beta_prior = 2
prior = beta(alpha_prior, beta_prior).pdf(x)

# Set a single random seed for reproducibility
np.random.seed(42)
true_p = 0.6  # True probability of heads

for sample_size in sample_sizes:
    flips = np.random.binomial(1, true_p, sample_size)
    observed_heads = np.sum(flips)
    posterior, mean, credible_interval = update_posterior(alpha_prior, beta_prior, observed_heads, sample_size - observed_heads)
    posteriors.append(posterior)
    means.append(mean)  
    credible_intervals.append(credible_interval)  

# Plotting the prior and updated posteriors for each sample size
plt.figure(figsize=(12, 6))
plt.plot(x, prior, label="Prior Distribution", color='blue')

# Iterate over each sample size and plot the corresponding posterior with its mean and credible interval in the label
for i, posterior in enumerate(posteriors):
    mean = means[i]
    ci_lower, ci_upper = credible_intervals[i]
    plt.plot(x, posterior, linestyle='--', label=f"Posterior after {sample_sizes[i]} tosses\n(mean = {mean:.3f}, 95% CI = [{ci_lower:.3f}, {ci_upper:.3f}])")

plt.xlabel("Probability of heads (θ)")
plt.ylabel("Density")
plt.title("Effect of Sample Size on Posterior Distribution")
plt.legend()
plt.show()
```

From this simple case, we can conclude:
1. **Frequentist Approach**:
   - Provides a single point estimate based on the data.
   - Does not incorporate prior beliefs or knowledge about the probability.
   - Limited in cases where prior information is valuable.

2. **Bayesian Approach**:
   - Combines prior beliefs with observed data, providing a posterior distribution.
   - Allows for a more flexible interpretation, showing the uncertainty around the estimate.
   - Particularly useful in scenarios with limited data or where prior knowledge is beneficial.

### The Effect of the Prior Distribution on Bayesian Inference

The prior distribution captures our beliefs about the parameter (in this case, the probability of heads) before we observe any data. Priors can be:
- **Informative Priors**: Reflect strong beliefs about the parameter's value (e.g., a biased coin).
- **Non-informative Priors**: Indicate little to no prior knowledge, often chosen to be neutral (e.g., a uniform distribution).

Let's see how different prior distributions affect the posterior distribution. We'll use three different priors:
1. **Uniform Prior**: Represents a non-informative belief, where we assume all probabilities are equally likely.
2. **Strong Prior Belief (Fair Coin)**: Represents a strong initial belief that the coin is fair (i.e., $\theta = 0.5$).
3. **Biased Prior**: Represents an initial belief that the coin is heavily biased towards heads.

Let's observe how each prior affects the posterior distribution after observing 20 coin tosses.


```python
# Define the true probability of heads and simulate the tosses
true_p = 0.6
sample_size = 20
np.random.seed(42)  # for reproducibility
flips = np.random.binomial(1, true_p, sample_size)
observed_heads = np.sum(flips)

# Defining three sets of prior parameters for comparison
priors = {
    "Uniform Prior": (1, 1),  # Non-informative prior
    "Strong Prior (Fair Coin)": (30, 30),  # Strong prior belief that theta = 0.5
    "Biased Prior (Heads)": (30, 5)  # Prior belief that the coin is biased towards heads
}

# Define the range for plotting the distributions
x = np.linspace(0, 1, 100)

# Loop through each prior and plot both the prior and posterior
for prior_name, (prior_alpha, prior_beta) in priors.items():
    # Calculate prior distribution
    prior = beta(prior_alpha, prior_beta).pdf(x)
    
    # Update posterior based on observed data
    posterior_alpha = prior_alpha + observed_heads
    posterior_beta = prior_beta + sample_size - observed_heads
    posterior = beta(posterior_alpha, posterior_beta).pdf(x)
    mean = posterior_alpha / (posterior_alpha + posterior_beta) 
    credible_interval = beta.ppf([0.025, 0.975], posterior_alpha, posterior_beta)
    
    # Plot prior and posterior distributions
    plt.figure(figsize=(10, 5))
    plt.plot(x, prior, label=f"Prior - {prior_name}", linestyle='-', color="blue", alpha=0.5)
    plt.plot(x, posterior, label=f"Posterior - {prior_name}\n(mean = {mean:.3f}, 95% CI = [{credible_interval[0]:.3f}, {credible_interval[1]:.3f}])", linestyle='-', color="orange")
    plt.xlabel("Probability of heads (θ)")
    plt.ylabel("Density")
    plt.title(f"{prior_name}: Prior vs. Posterior Distribution")
    plt.legend()
    plt.show()
```

# Conjugate Priors

In Bayesian inference, the term **conjugate prior** refers to a situation where the prior and posterior distributions are in the same family. This property simplifies calculations significantly, especially for complex models. Conjugate priors allow us to obtain a closed-form solution for the posterior distribution, making Bayesian updates efficient and computationally manageable.

### Why Use Conjugate Priors?
- **Simplifies Computation**: Conjugate priors yield a posterior distribution in the same family as the prior, allowing us to update beliefs without complex calculations.
- **Closed-Form Solutions**: In many cases, conjugate priors allow for a closed-form solution for the posterior, eliminating the need for numerical approximations.
- **Common in Bayesian Models**: Conjugate priors are widely used in Bayesian statistics due to their analytical convenience and interpretability.

Here are some common conjugate prior-likelihood pairs and their typical use cases:

### Common Conjugate Families

1. **Beta-Binomial Conjugate Pair**
   - **Prior**: Beta distribution
     - The Beta distribution, denoted as $\text{Beta}(\alpha, \beta)$, models our prior belief about the probability $p$ of a binary event (e.g., probability of heads in a coin flip).
     - **Expression**: $P(p) = \frac{p^{\alpha-1} (1 - p)^{\beta-1}}{B(\alpha, \beta)}$, where $B(\alpha, \beta)$ is the Beta function.

   - **Likelihood**: Binomial distribution
     - The Binomial distribution, denoted as $\text{Binomial}(n, p)$, models the likelihood of observing $k$ successes in $n$ trials, given the probability $p$ of success.
     - **Expression**: $P(k | p, n) = \binom{n}{k} p^k (1 - p)^{n - k}$

   - **Posterior**: Beta distribution
     - After observing $k$ successes in $n$ trials, the posterior distribution is updated as $\text{Beta}(\alpha + k, \beta + n - k)$.

   - **Use Case**: 
     - Useful in estimating probabilities in binary outcome scenarios, such as A/B testing and success/failure experiments.

2. **Gamma-Poisson Conjugate Pair (also known as Negative Binomial)**
   - **Prior**: Gamma distribution
     - The Gamma distribution, denoted as $\text{Gamma}(\alpha, \beta)$, models our prior belief about the rate $\lambda$ of a Poisson process (e.g., the rate of customer arrivals).
     - **Expression**: $P(\lambda) = \frac{\beta^{\alpha} \lambda^{\alpha-1} e^{-\beta \lambda}}{\Gamma(\alpha)}$, where $\Gamma(\alpha)$ is the Gamma function.

   - **Likelihood**: Poisson distribution
     - The Poisson distribution, denoted as $\text{Poisson}(\lambda)$, models the likelihood of observing $k$ events in a given time period, given the rate $\lambda$.
     - **Expression**: $P(k | \lambda) = \frac{\lambda^k e^{-\lambda}}{k!}$

   - **Posterior**: Gamma distribution
     - After observing during $n$ timesteps a total $k$ events, the posterior distribution for $\lambda$ is updated as $\text{Gamma}(\alpha + k, \beta + n)$.

   - **Use Case**:
     - Useful in count data modeling where events occur randomly over time, such as epidemiology, customer arrival rates, and reliability engineering.

3. **Normal-Normal Conjugate Pair (for Known Variance)**
   - **Prior**: Normal distribution
     - The Normal distribution, denoted as $\text{Normal}(\mu_0, \sigma^2_0)$, models our prior belief about the mean $\mu$ of a normally distributed variable.
     - **Expression**: $P(\mu) = \frac{1}{\sqrt{2 \pi \sigma_0^2}} e^{-\frac{(\mu - \mu_0)^2}{2 \sigma_0^2}}$

   - **Likelihood**: Normal distribution (with known variance)
     - The Normal distribution, denoted as $\text{Normal}(\mu, \sigma^2)$, represents the likelihood of observing data $x$ with mean $\mu$ and known variance $\sigma^2$.
     - **Expression**: $P(x | \mu) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}$

   - **Posterior**: Normal distribution
     - Given observed data $x$, the posterior distribution for $\mu$ is also a Normal distribution with updated parameters based on the prior mean $\mu_0$, prior variance $\sigma_0^2$, observed mean $\bar{x}$, and sample size $n$.

     - **Updated Mean**: $\mu_n = \frac{\sigma^2_0}{\sigma^2_0 + n \sigma^2} \bar{x} + \frac{n \sigma^2}{\sigma^2_0 + n \sigma^2} \mu_0$
     - **Updated Variance**: $\sigma^2_n = \frac{\sigma^2 \sigma^2_0}{n \sigma^2_0 + \sigma^2}$

   - **Use Case**:
     - Useful when estimating the mean of a normally distributed variable with known variance, often seen in quality control and sensor data analysis.

4. **Normal-Inverse Gamma Conjugate Pair (for Unknown Mean and Variance)**
   - **Prior**: Normal distribution for the mean and Inverse Gamma for the variance
     - The **Normal-Inverse Gamma distribution** is used as a joint prior for both the mean $\mu$ and the variance $\sigma^2$.
     - **Normal prior for $\mu$**: $P(\mu | \sigma^2) = \text{Normal}(\mu_0, \sigma^2 / \kappa_0)$
     - **Inverse Gamma prior for $\sigma^2$**: $P(\sigma^2) = \text{InvGamma}(\alpha_0, \beta_0)$

   - **Likelihood**: Normal distribution
     - Observed data $x$ follows a Normal distribution with mean $\mu$ and variance $\sigma^2$.
     - **Expression**: $P(x | \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}$

   - **Posterior**: Normal-Inverse Gamma distribution
     - After observing $n$ samples with mean $\bar{x}$ and variance $s^2$, the posterior distribution remains Normal-Inverse Gamma with updated parameters:
       - **Posterior Mean**: $\mu_n = \frac{\kappa_0 \mu_0 + n \bar{x}}{\kappa_0 + n}$
       - **Posterior Variance (Inverse Gamma parameters)**:
         - Updated $\alpha$: $\alpha_n = \alpha_0 + n/2$
         - Updated $\beta$: $\beta_n = \beta_0 + \frac{1}{2}(n s^2 + \frac{\kappa_0 n (\bar{x} - \mu_0)^2}{\kappa_0 + n})$

   - **Use Case**:
     - Useful for estimating both the mean and variance in normally distributed data when both parameters are unknown. Common in financial and environmental data analysis.

These conjugate pairs simplify the Bayesian inference process by providing closed-form solutions for the posterior distribution. By choosing appropriate conjugate priors, we can update our beliefs based on observed data.

However, while conjugate priors are extremely useful, they are not always available for every model or type of data. In these cases, Bayesian inference can still be performed using other methods, such as numerical approximation techniques (e.g., Markov Chain Monte Carlo, Variational Inference).

- The data is normally distributed, and
- We have a prior belief about the mean temperature with a known variance.

We want to estimate the average temperature in a city during summer. Based on previous years, the mean temperature was $33.5$ with a variance of $4.5$, which mean and standard deviation will be used to build our **prior**. We have observed daily temperatures over a period of days and use this **data** to update our belief about the average temperature, with a supposed variance of $5.0$.

### Model
- **Prior**: Normal distribution, representing our initial belief about the average temperature.
- **Likelihood**: Normal distribution, representing the observed data (daily temperatures).
- **Posterior**: Normal distribution, combining our prior belief with the observed data.
