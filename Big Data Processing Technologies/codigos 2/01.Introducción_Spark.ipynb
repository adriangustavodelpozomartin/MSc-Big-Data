## ¿Por qué Spark?
Hadoop nació (~2005) para procesar grandes cantidades de datos en paralelo. Poco a poco
fueron surgiendo nuevas problemáticas que no se podían resolver con el paradigma *MapReduce* y fueron apareciendo
nuevos proyectos para solventar estas
problemáticas, siendo necesario así una "jungla" de programas para un trabajo de big data:

&nbsp;    
&nbsp;    
&nbsp;    


![](img/mapreduce_ecosystem.png)


&nbsp;    
&nbsp;    
&nbsp;    


Spark nace (2009~2013) con dos finalidades claras: **ser rápido**, para ello cambia la manera de trabajar internamente (utiliza memoria, *RDD*,*DAG*...) y **unificar**
bajo un solo proyecto los grandes problemas de datos hasta el momento: Procesamiento en Batch,
en *streaming*, *machine learning*, *SQL*...   

Además incluye en el mismo proyecto varios lenguajes: Scala, Java, python y R.

<h1 style="font-size:40px;">Conceptos Básicos</h1>

Para ayudarno a entender los principales conceptos tenemos que tener muy presentes sus principales ventajas:

* Mayor flexibilidad en la definición de transformaciones
* Menor uso de almacenamiento en disco
* Aprovechamiento de la memoria
* Tolerancia a fallos
* API muy expresiva (inspirada en Scala)

## `RDD` (*Resilient Distributed Datasets*)

* Abstracción básica en Spark
* Contiene las transformaciones que se van a realizar sobre un conjunto de datos, propiedades:
    * Inmutable
    * Lazy evaluation
    * En caso de fallo se puede recuperar el estado
    * Control de persistencia y particionado

##  `SparkContext`
* Representa la conexión a un cluster Spark
* Permite crear distintos tipos de variables
    * RDD
    * DataFrame
    * Acumuladores
    * Broadcast

##  Task Scheduler (Planificador de tareas)
* Prepara los *jobs* a traves de DAG (*Directed Acyclic Graph*).
* Los jobs se separan en *stages*.
* Agrupa funciones para ejecutar los *stages*.
* *Cache-aware*: tienen en cuenta la utilización y localización (data locality) de los datos.
* *Partitioning-aware*: tiene en cuenta la partición de los datos para reducir shuffles por la red.

<h1 style="font-size:40px;">API: Transformaciones vs Acciones</h1>

Otra ventaja de Spark es que tiene una API intuitiva, muy expresiva que permitece una gran flexibilidad. Veamos un ejemplo del típico "*Word Count*":

![](https://slideplayer.com/slide/14330321/89/images/14/Code%3A+Hadoop+vs+Spark+%28e.g.%2C+Word+Count%29.jpg)

En Spark, podemos diferencias dos tipos de operaciones:

* **Transformaciones**: Construyen un nuevo RDD a partir de uno creado previamente. **Estas transformaciones son lazy**.
* **Acciones**: Ejecuta una **acción final** basándose en un RDD existente: imprimir por pantalla, contar, escribir... (No son lazy)


![](img/RDD_Operations.png)

Lo mejor es consultar directamente la documentación oficial:

* https://spark.apache.org/docs/2.4.0/rdd-programming-guide.html#transformations
* https://spark.apache.org/docs/2.4.0/rdd-programming-guide.html#actions

### Word Count con `pyspark`

````python
readme = spark.sparkContext.textFile("/datos/README.md")
````

````python
readme.take(1)
````

````python
readme.count()
````

````python
import re
````

````python
(
    readme
    # Divido cada por espacios
    .flatMap(lambda x: re.split(r"\s+", x))
    # Creo un pair RDD
    .map(lambda x: (x, 1))
    # Reduzco por key y sumo los unos para contar
    .reduceByKey(lambda a, b: a + b)
    # Ordeno de mayor a menor el conteo
    .sortBy(lambda x: -x[1])
).take(10)
````

### De python a spark

Podemos pasar variables de python directamente a *RDD's* y viceversa:

````python
import numpy as np
````

````python
colores = np.array(["blue", "red", "green", "yellow", "brown", "black"])
````

````python
muestra = np.random.choice(colores, 1000000)
````

````python
len(muestra)
````

````python
muestra[:10]
````

````python
rdd_muestra = spark.sparkContext.parallelize(muestra, 10)
````

````python
rdd_muestra.count()
````

````python
rdd_muestra.take(10)
````

````python
(rdd_muestra.map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b)).collect()
````

````python
np.unique(muestra, return_counts=True)
````

¿Y si los datos no caben en memoría?

````python
gran_muestra = (
    spark.sparkContext
    # Usamos range como si fuera un bucle
    .range(10, numSlices=200).flatMap(lambda _: np.random.choice(colores, 1000000))
)
````

````python
gran_muestra.take(10)
````

````python
print("%.2E" % (10 * 1000000))
````

````python
n = gran_muestra.count()
n
print("%.2E" % n)
````

````python
(
    gran_muestra
    # Map
    .map(lambda x: (x, 1))
    # Reduce
    .reduceByKey(lambda a, b: a + b)
).collect()
````

### <font color='darkred'> CUIDADO:

Al usar `collect` se recoge todo el *RDD* en el driver, así que hay que estar seguro de que el tamaño sea pequeño o tendremos problemas de memoria.

`
np.unique(gran_muestra.collect(),return_counts=True)
`

### Transformaciones y Acciones

La API de Pyspark es muy parecida al core en *scala*. También tenemos transformaciones y acciones:    
&nbsp;   

![](img/RDD_Operations.png)

* https://spark.apache.org/docs/latest/rdd-programming-guide.html#transformations

````python
peliculas = [
    "http://www.imdb.com/title/tt0071562/reference",
    "http://www.imdb.com/title/tt0110912/reference",
    "http://www.imdb.com/title/tt0050083/reference",
    "http://www.imdb.com/title/tt0108052/reference",
    "http://www.imdb.com/title/tt0468569/reference",
    "http://www.imdb.com/title/tt0068646/reference",
    "http://www.imdb.com/title/tt0167260/reference",
    "http://www.imdb.com/title/tt0060196/reference",
    "http://www.imdb.com/title/tt0137523/reference",
    "http://www.imdb.com/title/tt0111161/reference",
]
````

````python
import requests
from bs4 import BeautifulSoup
````

````python
def parsear_html(texto):
    soup = BeautifulSoup(texto, "lxml")
    item = {}
    item["titulo"] = soup.find("h3").find(text=True).replace("\xa0", " ").strip()
    item["ratingvalue"] = float(soup.select_one(".ipl-rating-star__rating").text)
    return item
````

````python
descargas = spark.sparkContext.parallelize(peliculas).map(lambda x: requests.get(x).content)
````

````python
descargas.count()
````

````python
parseados = descargas.map(parsear_html)
````

````python
parseados.collect()
````

### Cache

Podemos cachear un *RDD*, para no tener que recalcuarlo cada vez. Muy útil cuando estamos explorando los datos o tenemos que hacer dos acciones distintas sobre el mismo *RDD*.

````python
%%timeit
descargas.count()
````

````python
%%timeit
parseados.collect()
````

````python
descargas.cache()
````

````python
%%timeit
descargas.count()
````

````python
parseados = descargas.map(parsear_html)
````

````python
%%timeit
parseados.collect()
````

````python
descargas.is_cached
````

````python
descargas.unpersist()
````

````python
descargas.is_cached
````
