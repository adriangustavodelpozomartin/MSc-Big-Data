```python
import os
import requests

# Parámetros de Entrada
input_folder = "urls_por_fecha"  # Carpeta con los archivos de URLs
output_folder = "pdfs_actos_inscritos"  # Carpeta donde se guardarán los PDFs

# 📁 Crear Carpeta para Almacenar los PDFs
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"✅ Carpeta '{output_folder}' creada correctamente.")
else:
    print(f"📂 Carpeta '{output_folder}' ya existe.")

# 🔍 Verificar que la Carpeta de URLs Existe
if not os.path.exists(input_folder):
    print(f"❌ La carpeta '{input_folder}' no existe. Asegúrate de ejecutar la fase Spider primero.")
    exit(1)

# 📂 Procesar Cada Archivo de URLs
try:
    archivos_urls = [f for f in os.listdir(input_folder) if f.endswith(".txt")]
    
    if not archivos_urls:
        print(f"❌ No se encontraron archivos de URLs en la carpeta '{input_folder}'.")
        exit(1)
    
    print(f"✅ Se han encontrado {len(archivos_urls)} archivos de URLs para procesar.")
    
    for archivo in archivos_urls:
        archivo_path = os.path.join(input_folder, archivo)
        fecha = os.path.splitext(archivo)[0]  # Extraer la fecha del nombre del archivo

        print(f"\n📅 Procesando archivo: {archivo} (Fecha: {fecha})")
        
        # Crear una subcarpeta para cada fecha dentro de output_folder
        fecha_folder = os.path.join(output_folder, fecha)
        if not os.path.exists(fecha_folder):
            os.makedirs(fecha_folder)
            print(f"✅ Carpeta creada para la fecha '{fecha}': {fecha_folder}")
        
        # Leer las URLs del archivo
        with open(archivo_path, "r") as file:
            urls = file.readlines()
            urls = [url.strip() for url in urls if url.strip()]
        
        print(f"🔗 Se han encontrado {len(urls)} URLs para descargar en '{archivo}'.")

        # Descargar Cada PDF
        for index, url in enumerate(urls):
            try:
                response = requests.get(url, stream=True)
                response.raise_for_status()  # Lanza una excepción si hay un error en la solicitud

                # Extraer el nombre del archivo PDF de la URL
                filename = os.path.join(fecha_folder, os.path.basename(url))
                
                # Guardar el PDF en la subcarpeta correspondiente
                with open(filename, "wb") as pdf_file:
                    for chunk in response.iter_content(chunk_size=1024):
                        pdf_file.write(chunk)
                
                print(f"✅ [{index+1}/{len(urls)}] PDF descargado: {filename}")
            
            except requests.exceptions.RequestException as e:
                print(f"❌ Error al descargar {url}: {e}")
            except Exception as e:
                print(f"❌ Error inesperado al procesar {url}: {e}")

except Exception as e:
    print(f"❌ Error inesperado: {e}")
```
