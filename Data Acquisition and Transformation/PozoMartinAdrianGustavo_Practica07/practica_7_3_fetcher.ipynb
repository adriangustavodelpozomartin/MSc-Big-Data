```python
# Importamos las librerías necesarias
import csv
import requests
from bs4 import BeautifulSoup
import csv
import requests
import time
import os
```

```python
# Creamos una carpeta llamada "university_htmls" para guardar los archivos HTML de las universidades.
# Si la carpeta ya existe, con exist_ok=True nos aseguramos que no hacemos nada.
output_dir = "university_htmls"
os.makedirs(output_dir, exist_ok=True)

# Abrimos el archivo "universities.csv" en modo lectura para leer los datos de las universidades.
with open("universities.csv", "r", encoding="utf-8") as csvfile:
    # Creamos un lector de archivos CSV, configurado para usar el punto y coma como delimitador.
    reader = csv.reader(csvfile, delimiter=';')
    # Saltamos la primera fila del archivo (la cabecera), ya que contiene solo los nombres de las columnas.
    next(reader)

    # Iteramos sobre las filas restantes del archivo, que contienen los datos de las universidades.
    for row in reader:
        # Extraemos el nombre del país, la URL del país, el nombre de la universidad y la URL de la universidad.
        country_name, country_url, university_name, university_url = row

        # Creamos un nombre de archivo seguro para guardar el HTML de cada universidad.
        # Reemplazamos espacios y caracteres no válidos como "/" con guiones bajos para evitar errores.
        sanitized_university_name = university_name.replace(" ", "_").replace("/", "_")
        # Construimos la ruta completa del archivo donde se guardará el HTML. Es decir, en el directorio university_htmls creamos un archivo
        # con el nombre de la universidad.
        # Por ejemplo, si la universidad es "University of Berlin" y la carpeta de salida es "university_htmls":
        # - sanitized_university_name será "University_of_Berlin" (los espacios se reemplazan por guiones bajos).
        # - output_file será "university_htmls/University_of_Berlin.html", lo que significa que el archivo se guardará
        #   en la carpeta "university_htmls" con el nombre "University_of_Berlin.html".
        output_file = os.path.join(output_dir, f"{sanitized_university_name}.html")

        # Si el archivo ya existe en la carpeta, lo saltamos para evitar volver a descargarlo. (Nos permite ahorrar tiempo en las pruebas)
        if os.path.exists(output_file):
            print(f"Archivo ya descargado: {output_file}")
            continue
        # El manejo de excepciones va a ser importante ya que, como veremos hay casos en los que se lanza un error. Por ejemplo,
        # vamos a ver como obtenemos un error 404 para la Hochschule Bremen o la Brandenburg Medical School Theodor Fontane.
        try:
            # Realizamos una solicitud HTTP para descargar el contenido HTML de la página de la universidad.
            response = requests.get(university_url)
            # Verificamos si la solicitud fue exitosa antes de guardar el contenido.
            if response.status_code == 200:
                # Guardamos el contenido HTML en el archivo especificado, utilizando codificación UTF-8.
                with open(output_file, "w", encoding="utf-8") as file:
                    file.write(response.text)
                # Informamos que el archivo se descargó correctamente.
                print(f"Descargado: {university_name} -> {output_file}")
            else:
                # Informamos si hubo un error al descargar, incluyendo el código de estado HTTP.
                print(f"Error al descargar {university_name}: Código {response.status_code}")
        except Exception as e:
            # Informamos si ocurrió un error inesperado durante la descarga o el guardado. Es necesario hacer esto ya que, como vemos, 
            # en el output, obtenemos un error 404 para la Hochschule Bremen o la Brandenburg Medical School Theodor Fontane.
            print(f"Error al procesar {university_name}: {e}")

        # Añadimos una pausa de 1 segundo entre descargas para no sobrecargar el servidor web.
        time.sleep(1)

# Informamos que todas las descargas han finalizado.
print("Descargas completadas.")                    
```
