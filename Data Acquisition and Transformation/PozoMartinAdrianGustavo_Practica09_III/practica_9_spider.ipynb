# PRÁCTICA 9. SPIDER

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time
import csv
```

```python
# Ahora inicializamos el navegador Chrome que será controlado por Selenium.
# Usamos 'webdriver.Chrome()' para lanzar una nueva instancia del navegador Chrome.
driver = webdriver.Chrome()

# Ahora configuramos el navegador para que se abra en modo pantalla completa.
# Esto es útil porque nos permite ver toda la página y evitar problemas con elementos que se ocultan cuando no están en pantalla completa
# (por ejemplo,la primera de las tablas de las que extraemos la información)
driver.maximize_window()

# Ahora cargamos la URL de la página que queremos automatizar.
# Usamos 'driver.get()' para acceder a la página web.
url = "https://comparador.cnmc.gob.es/"
driver.get(url)
```

## 1.- INTRODUCIMOS LOS PARÁMETROS DE MANERA AUTOMÁTICA

### 1.1- PRIMERA PESTAÑA

```python
### Paso 1: Cerramos el pop-up de las cookies
try:
    # Ahora buscamos el botón para rechazar las cookies usando su clase CSS.
    # 'driver.find_element(By.CLASS_NAME, "cookiesjsr-btn")' identifica el botón en la página.
    boton_denegar = driver.find_element(By.CLASS_NAME, "cookiesjsr-btn")

    # Ahora hacemos clic en el botón para cerrar el pop-up.
    # Si el botón de cookies se encuentra correctamente, el pop-up desaparecerá.
    boton_denegar.click()
    print("Pop-up de cookies cerrado correctamente.")

    # Añadimos una pausa de 2 segundos para dar tiempo a que la acción sea procesada por el navegador.
    time.sleep(2)
except Exception as e:
    # Si algo falla (por ejemplo, si no se encuentra el botón), capturamos el error y mostramos un mensaje.
    # Si el botón no existe, imprimiremos "No se pudo cerrar el pop-up de cookies: {e}".
    print(f"No se pudo cerrar el pop-up de cookies: {e}")

    # Paso 2: Seleccionar "Electricidad" y aceptar
try:
        
        # Paso 2.1: Hacemos clic en el slot del menú desplegable
    try:
        # Paso 1: Localizamos y hacemos clic en el menú desplegable
        menu_desplegable = driver.find_element(By.XPATH, '//label[text()="Tipo de suministro que desea contratar:"]/following-sibling::input')
        menu_desplegable.click()
        print("Menú desplegable relacionado con el label identificado y seleccionado correctamente.")
        time.sleep(5)  # Pausa para permitir que el menú se cargue completamente
        
        # El presente siguiente ha requerido la incorporación de métodos alternativos, debido a que las 
        # funciones empleadas inicialmente generaban errores en la ejecución. En concreto, se intentó localizar y seleccionar el elemento 
        # correspondiente a la opción "Electricidad" mediante el uso de las funciones `find_element` con expresiones XPath, como:
        # opcion_electricidad = driver.find_element(By.XPATH, '//div[@class="v-list-item__title" and text()="Electricidad"]')
        # Aunque válida, esta estrategia resultó en un error  (`has no size and location`), ya que Selenium no podía 
        # interactuar con el elemento, posiblemente debido a su estado no visible o a que estaba parcialmente renderizado en el DOM.
        # Ante estos inconvenientes, se optó por implementar un enfoque basado en el uso del atributo `id`, combinado con la función 
        # `execute_script` de Selenium. Este método permitió garantizar la interacción con el elemento forzando el desplazamiento 
        # (`scrollIntoView`) y la ejecución de un clic mediante JavaScript. Este enfoque ha demostrado ser eficaz para superar los problemas 
        # previos y seleccionar correctamente la opción deseada en el menú desplegable.

        opcion_electricidad = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, 'list-item-66-0'))
        )
    
        # Paso 3: Asegurarnos de que el elemento es visible
        driver.execute_script("arguments[0].scrollIntoView(true);", opcion_electricidad)
    
        # Paso 4: Hacer clic en la opción usando JavaScript para garantizar interacción
        driver.execute_script("arguments[0].click();", opcion_electricidad)
        print("Opción 'Electricidad' seleccionada correctamente.")
        time.sleep(2)
        
    except Exception as e:
            # Capturamos errores y mostramos un mensaje para diagnosticar problemas
            print(f"Error al seleccionar la opción 'Electricidad': {e}")



    # Ahora buscamos el botón "Iniciar" usando su texto parcial para identificarlo.
    # 'By.XPATH' localiza el botón que contiene el texto "Iniciar".
    boton_aceptar = driver.find_element(By.XPATH, '//span[contains(text(), "Iniciar")]')

    # Hacemos clic en el botón "Iniciar" utilizando un script de JavaScript.
    # Esto tuvimos que hacerlo porque obteníamos error ya que no conseguíamos que Selenium interactuara directamente con el botón.
    driver.execute_script("arguments[0].click();", boton_aceptar)
    print("Botón 'Iniciar' encontrado y presionado.")

    # Pausa de 2 segundos para dar tiempo a que la acción tenga efecto.
    time.sleep(2)
except Exception as e:
    # Si algo falla durante este proceso, capturamos el error y mostramos un mensaje.
    # Ejemplo: Si no se encuentra el botón "Iniciar", imprimiremos el mensaje de error.
    print(f"Hubo un problema al seleccionar 'Electricidad' o al intentar presionar 'Iniciar': {e}")
```

### 1.2.- SEGUNDA PESTAÑA

```python
# Ahora vamos a interactuar con el campo de código postal
try:
    # Esperamos hasta que el campo de entrada para el código postal esté presente en la página.
    # Usamos 'WebDriverWait' para esperar hasta 10 segundos antes de que ocurra un timeout. Hemos tenido que usar esta función ya que 
    # con driver.find_element no conseguíamos encontrar el slot del Código Postal.
    # 'EC.presence_of_element_located' verifica que el campo esté cargado y listo para ser usado.
    # Si el campo tarda en aparecer, este método garantiza que no intentemos interactuar antes de tiempo.
    input_field = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//input[@name="codigoPostal" and @maxlength="5"]'))
    )
    print("Campo de código postal localizado correctamente.")

    # Ahora ingresamos el código postal en el campo de entrada.
    # Asignamos el valor "37185" a la variable 'postal_code' y lo enviamos al campo con 'send_keys'.
    postal_code = "28008"
    input_field.send_keys(postal_code)
    print("Código postal ingresado correctamente.")
except Exception as e:
    # Capturamos cualquier error que ocurra, como que el campo no sea encontrado, y mostramos un mensaje.
    # Ejemplo: Si la página no carga correctamente, podría aparecer un error como "TimeoutException".
    print(f"Error al localizar o interactuar con el campo de código postal: {e}")

# Ahora buscamos y hacemos clic en el botón "Continuar"
try:
    # Localizamos el botón "Continuar" usando su texto visible.
    # Con'By.XPATH' podemos buscar cualquier elemento cuyo texto parcial sea "Continuar".
    boton_continuar = driver.find_element(By.XPATH, '//span[contains(text(), "Continuar")]')
    print("Botón 'Continuar' localizado correctamente.")

    # Hacemos clic en el botón "Continuar" usando el método 'click()'.
    # Esto simula la acción de presionar el botón con el ratón.
    # Esto iniciará la acción siguiente en la página.
    boton_continuar.click()
    print("Acción realizada: Se presionó el botón 'Continuar'.")

    # Pausa de 2 segundos para garantizar que la acción sea procesada antes de continuar con el siguiente paso.
    time.sleep(2)
except Exception as e:
    # Si no encontramos el botón o no podemos interactuar con él, capturamos el error y mostramos un mensaje.
    # Ejemplo: Si el botón está deshabilitado, podría lanzarse una excepción como "ElementNotInteractableException".
    print(f"No se pudo interactuar con el botón 'Continuar': {e}")
```

## 2.- SPIDER. EXTRACCIÓN DE LOS DATOS

```python


# Paso 1: Localización de la tabla en la página
try:
    # Ahora identificamos la tabla HTML en la página usando un selector XPath.
    # La tabla que buscamos está ubicada dentro de un elemento con id 'listadoOfertas'.
    # Ejemplo: Si el HTML contiene:
    # <div id="listadoOfertas">
    #     <table>...</table>
    # </div>
    # El XPath '//*[@id="listadoOfertas"]/div/table' selecciona esa tabla.
    tabla = driver.find_element(By.XPATH, '//*[@id="listadoOfertas"]/div/table')
    time.sleep(2)  # Añadimos una pausa para evitar errores al procesar elementos dinámicos.
except Exception as e:
    # Si no podemos encontrar la tabla, mostramos un error y cerramos el programa.
    print(f"Error al encontrar la tabla: {e}")
    driver.quit()
    exit()

# Paso 2: Extraección de las filas de la tabla
# Ahora obtenemos todas las filas (<tr>) dentro de la tabla.
# Esto incluye tanto las filas de encabezado como las filas de datos.
filas = tabla.find_elements(By.XPATH, ".//tr")
time.sleep(2)  # Pausa para evitar sobrecargar el servidor.

# Creamos una lista vacía para almacenar los datos extraídos de la tabla.
datos = []

# Paso 3: Procesamiento de cada fila de la tabla
for fila in filas[2:]:  # Omitimos las primeras dos filas ya que son cabeceras.
    # Localizamos las celdas (<td>) dentro de cada fila.
    celdas = fila.find_elements(By.TAG_NAME, "td")
    time.sleep(1)  # Añadimos una pausa para evitar cargar demasiados datos de golpe.

    # Verificamos si la fila tiene celdas. Si está vacía, pasamos a la siguiente fila.
    if len(celdas) == 0:
        continue

    # Creamos una lista temporal para almacenar los datos de esta fila.
    datos_fila = []

    # Paso 3.1: Extraección del nombre de la comercializadora desde la imagen en la primera celda.
    try:
        # Buscamos la etiqueta <img> dentro de la primera celda y obtenemos el atributo 'alt'.
        # Ejemplo: Si el HTML contiene:
        # <td><img alt="Endesa" /></td>
        # El valor de 'empresa_comercializadora' será "Endesa".
        img_element = celdas[0].find_element(By.TAG_NAME, "img")
        empresa_comercializadora = img_element.get_attribute("alt")
    except Exception:
        # Si no encontramos la imagen, asignamos "No disponible".
        empresa_comercializadora = "No disponible"
    datos_fila.append(empresa_comercializadora)

    # Paso 3.2: Extraección de texto de las celdas restantes, excepto las dos últimas.
    for i, celda in enumerate(celdas[1:-2]):
        # Obtenemos el texto de la celda y eliminamos espacios innecesarios.
        texto_celda = celda.text.strip()

        # Para los campos Importe primeras facturas en Euros e Importe siguientes facturas en Euros, adaptamos lo capturado para que sea
        # flotante.
        # Así, si el índice de la celda es 2 o 3, convertimos el valor a un número flotante.
        if i in [2, 3]:
            try:
                # Eliminamos el símbolo de euro (€) y convertimos el formato capturado (1.198,65) a (1198.65).
                texto_celda = texto_celda.replace("€", "").strip()
                texto_celda = texto_celda.replace(".", "").replace(",", ".")
                texto_celda = float(texto_celda)
            except ValueError:
                # Si no podemos convertir el valor, asignamos 'None'.
                texto_celda = None
        datos_fila.append(texto_celda)

    # Paso 3.3: Extraección del penúltimo campo basado en el color del icono.
    try:
        # Buscamos un ícono (<i>) en la penúltima celda y obtenemos su clase CSS.
        # Ejemplo: Si el HTML contiene:
        # <i class="green--text"></i>, el valor será "Sí". Si <i class="grey--text"></i>, el valor será "No"
        penultimo_campo_icon = celdas[-2].find_element(By.TAG_NAME, "i")
        penultimo_campo_clase = penultimo_campo_icon.get_attribute("class")
        if "green--text" in penultimo_campo_clase:
            penultimo_campo = "Sí"
        elif "grey--text" in penultimo_campo_clase:
            penultimo_campo = "No"
        else:
            penultimo_campo = "Desconocido"
    except Exception:
        penultimo_campo = "No disponible"
    datos_fila.append(penultimo_campo)

    # Paso 3.4: Extraección del enlace del botón "ojo" en la última celda.
    try:
        # Buscamos un enlace (<a>) en la última celda y obtenemos su atributo 'href'.
        enlace_element = celdas[-1].find_element(By.TAG_NAME, "a")
        ultimo_campo = enlace_element.get_attribute("href")
    except Exception:
        # Si no encontramos el enlace, asignamos "No disponible".
        ultimo_campo = "No disponible"
    datos_fila.append(ultimo_campo)

    # Añadimos los datos de esta fila a la lista principal de datos.
    datos.append(datos_fila)

# Cerramos el navegador una vez que terminamos de extraer los datos.
driver.quit()

# Paso 4: Guardamos los datos que hemos extraídos en un archivo CSV
# Definimos las cabeceras para el archivo CSV.
cabeceras = [
    "Denominación Social Empresa Comercializadora",
    "Denominación de la Oferta",
    "Tipo de tarifa",
    "Importe primeras facturas en Euros",
    "Importe siguientes facturas en Euros",
    "Validez",
    "Servicios Adicionales Incluidos",
    "Penalización",
    "Verde",
    "URL de la Oferta"
]

# Abrimos un archivo CSV en modo escritura y escribimos los datos extraídos.
with open("datos_ofertas.csv", mode="w", newline="", encoding="utf-8") as file:
    # Creamos un objeto escritor CSV y definimos ';' como separador.
    writer = csv.writer(file, delimiter=";")
    # Escribimos las cabeceras en la primera fila.
    writer.writerow(cabeceras)
    # Escribimos las filas de datos extraídos.
    writer.writerows(datos)

# Imprimimos un mensaje de éxito en caso de que todo el proceso se haya completado correctamente.
print("Datos guardados en 'datos_ofertas.csv'")
```
